```python
import sub_logic
from matplotlib import pyplot as plt

if __name__ == "__main__":
    only_good = sub_logic.SubLog(False, {'a': 0.5, 'r': 1, 's':0})
    only_good2 = sub_logic.SubLog(False, {'a': 0.9, 'r': 1, 's':0})
    only_good3 = sub_logic.SubLog(False, {'a': 0.1, 'r': 1, 's':0})
    only_bad = sub_logic.SubLog(False, {'a': 0.5, 'r': 0, 's':1})
    only_bad2 = sub_logic.SubLog(False, {'a': 0.9, 'r': 0, 's':1})
    only_bad3 = sub_logic.SubLog(False, {'a': 0.1, 'r': 0, 's':1})
    good_bad = sub_logic.SubLog(False, {'a': 0.5, 'r': 1, 's':1})
    good_bad2 = sub_logic.SubLog(False, {'a': 0.9, 'r': 1, 's':1})
    good_bad3 = sub_logic.SubLog(False, {'a': 0.1, 'r': 1, 's':1})
    mygraph = sub_logic.Graph('car1')
```

```python
mygraph.add_direct_nodes('car2.5', only_good)
mygraph.add_direct_nodes('car2.8', only_good2)
mygraph.add_direct_nodes('car2.2', only_good3)
print(mygraph.get_my_opinion('car2.5').trust())
print(mygraph.get_my_opinion('car2.8').trust())
print(mygraph.get_my_opinion('car2.2').trust())
```

```python
mygraph.add_direct_nodes('car3.5', only_bad)
mygraph.add_direct_nodes('car3.8', only_bad2)
mygraph.add_direct_nodes('car3.2', only_bad3)
print(mygraph.get_my_opinion('car3.5').trust())
print(mygraph.get_my_opinion('car3.8').trust())
print(mygraph.get_my_opinion('car3.2').trust())
```

```python
mygraph.add_direct_nodes('car4.5', good_bad)
mygraph.add_direct_nodes('car4.8', good_bad2)
mygraph.add_direct_nodes('car4.2', good_bad3)
print(mygraph.get_my_opinion('car4.5').trust())
print(mygraph.get_my_opinion('car4.8').trust())
print(mygraph.get_my_opinion('car4.2').trust())
```

```python
for i in range(50):
    if i%2 == 0:
        new_data = sub_logic.SubLog(False, {'r' : 1, 's': 0, 'a' : 0.5})
    else:
        new_data = sub_logic.SubLog(False, {'r' : 0, 's': 1, 'a' : 0.5})
    mygraph.add_direct_nodes('car4.5', new_data)
    plt.rcParams["figure.figsize"] = [7.00, 3.50]
    plt.rcParams["figure.autolayout"] = True
    x = [i+1]
    y = [mygraph.get_my_opinion('car4.5').trust()]
    plt.xlim(0, 50)
    plt.ylim(0, 1)
    plt.grid()
    plt.plot(x, y, marker="o", markersize=4, markeredgecolor="white", markerfacecolor="red")
plt.title("Trust levels with good and bad interactions and base rate 0.5")
plt.xlabel("Number of interactions")
plt.ylabel("Trust level")
plt.savefig('gb05.png')
print('# good interactions '+str(round(mygraph.get_my_opinion('car4.5').r)))
print('# bad interactions '+str(round(mygraph.get_my_opinion('car4.5').s)))
```
